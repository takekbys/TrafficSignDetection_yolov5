```python
import json
```

```python
def load_key(keyname):
    
    key = open(keyname)
    key = json.load(key)
    
    key_inv = {v:k for k, v in key.items()}
    return key_inv

mapillary_key = load_key("MAPILLARY_TS_label_map.json")
mapillary_key
```

```python
def mapillary_to_yolo(annoname, key):
    annotation = open(annoname)
    annotation = json.load(annotation)
    
    img_width = annotation["width"]
    img_height = annotation["height"]
    objects = annotation["objects"]
    object_yolo = []
    
    for obj in objects:
        label = obj["label"]
        xmin = obj["bbox"]["xmin"]
        xmax = obj["bbox"]["xmax"]
        ymin = obj["bbox"]["ymin"]
        ymax = obj["bbox"]["ymax"]
        
        label_yolo = key[label]
        c_x = (xmin+xmax)/2/img_width
        c_y = (ymin+ymax)/2/img_height
        w = (-xmin+xmax)/img_width
        h = (-ymin+ymax)/img_height
        
        annotation_txt = "{} {:.6f} {:.6f} {:.6f} {:.6f}".format(label_yolo, c_x, c_y, w, h)
        object_yolo.append(annotation_txt)
        
    return object_yolo
```

```python
filename = "__n7yWAdeWtVB_N968KgbA.json"
mapillary_to_yolo(filename, mapillary_key)
```

```python
json_path = ""
text_path = ""
```

```python
def save_annotation(annoname):
    
    json_file = json_path+annoname+".json"
    text_file = text_path+annoname+".txt"
    
    annotation_yolo = mapillary_to_yolo(json_file, mapillary_key)
    
    with open(text_file, "w") as f:
        for anno in annotation_yolo:
            f.write(anno)
            f.write("\n")

save_annotation("sample")
```

```python
# validation dataの変換
json_list = "../mtsd_fully_annotated_annotation/mtsd_v2_fully_annotated/splits/val.txt"
json_path = "../mtsd_fully_annotated_annotation/mtsd_v2_fully_annotated/annotations/"
text_path = "../labels/val/"

with open(json_list, "r") as f:
    lines = f.read().splitlines()
    for annoname in lines:
        save_annotation(annoname)
```

```python
# train dataの変換
json_list = "../mtsd_fully_annotated_annotation/mtsd_v2_fully_annotated/splits/train.txt"
json_path = "../mtsd_fully_annotated_annotation/mtsd_v2_fully_annotated/annotations/"
text_path = "../labels/train/"

with open(json_list, "r") as f:
    lines = f.read().splitlines()
    for annoname in lines:
        save_annotation(annoname)
```
